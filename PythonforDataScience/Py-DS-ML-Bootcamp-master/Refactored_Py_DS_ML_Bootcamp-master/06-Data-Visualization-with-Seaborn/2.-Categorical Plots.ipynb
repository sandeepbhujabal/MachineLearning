```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='tip',data=tips)
```

```python
sns.barplot(x='sex',y='tip',data=tips,estimator=np.sum)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.sum)
```

```python
sns.countplot(x='smoker',data=tips)
```

```python
sns.boxplot(y='total_bill',x='day',data=tips)
```

```python
sns.boxplot(y='total_bill',x='day',data=tips, hue='sex')
```

```python
sns.violinplot(y='total_bill',x='day',data=tips)
```

```python
sns.stripplot(y='total_bill',x='day',data=tips)
```

```python
sns.swarmplot(y='total_bill',x='day',data=tips)
```

```python
sns.violinplot(y='total_bill',x='day',data=tips)
sns.swarmplot(y='total_bill',x='day',data=tips,color='black')
```

```python
sns.factorplot(y='total_bill',x='day',data=tips)
```

```python
sns.factorplot(y='total_bill',x='day',data=tips,kind='bar')
```

```python
sns.factorplot(y='total_bill',x='day',data=tips,kind='violin')
```

```python
sns.factorplot(y='total_bill',x='day',data=tips,kind='strip')
```

```python
sns.factorplot(y='total_bill',x='day',data=tips,kind='box')
```
